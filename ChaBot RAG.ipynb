```python
# Summary of Code by Pedro Yanez:
# It allows to query your own big corpus, using query api.
# Install libraries
# Import langchain, embedding and vectorizing functions
# Indicate openai personal key
# 959 pages pdf "Handbook of Conflict Resolution" is transformed to corpus, as .txt, using the webpage https://www.sketchengine.eu/
# communicate and allow google colab to access my google drive
# load pdf. First loaded an extract then tried to load entire file. Through trial and error I realized: I can only load small files, beause of token restriction
# to embeddings when files exceed certain amount of tokens.
# Also I realized is not possible to vectorize more than two files per minute because of restriction related to tokens allowed to vectorize per minute.
# For this reason, the big file Handbook of Conflict Resolution.txt, is converted to many small files. Each file has 300 lines and the files
# are called small file 300 , small file 600, small file 900, and so on, until small file 38700.
# each file , doing a for loop starting in 300 to 600  to 900 and so on, are loaded and then vectorized. There is a sleep of 30 seconds, to make sure only
# two files per minute and loaded and vectorized. In this way, it doesn't fail due to restrictions related to size of file, or amount of files to vectorize
# per minute.
# The query "who is ellen raider", answers "I don't know" before the loading of the entire corpus.
# After the loading of the corpus, the query brings the answer.
```
